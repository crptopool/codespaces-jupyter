```python
import fnmatch
### Search for all 5 letter words from dictionary containing all of input letters in any position ###

# from asyncio import loop
import urllib.request

with urllib.request.urlopen("https://raw.githubusercontent.com/charlesreid1/five-letter-words/master/sgb-words.txt") as url:
  sl = url.read().decode()
  words = sl.splitlines()

stripped_wordlist = [s.strip() for s in words]
stripped_wordlist1 = list(map(str.strip, words))

import pandas as pd

df = pd.DataFrame({'Wordlist': stripped_wordlist1})

df1 = pd.DataFrame()

df1[['0', 'Letter1', 'Letter2', 'Letter3', 'Letter4', 'Letter5', '6']] = df['Wordlist'].str.split(pat = '', expand=True)


# Remove three columns as index base
df1.drop(df1.columns[[0, 6]], axis=1, inplace=True)
df1['wordlist'] =  stripped_wordlist1
#set column as index
df1 = df1.set_index('wordlist')

# alphabet_count_array = [26][5]

import numpy as np

# create a 2D array with shape (3, 4)
# shape = (5, 26)
arr =  np.arange(1,131,1).reshape(5, 26)
# print(arr)

import string
lower = string.ascii_lowercase
list_of_lowercase_letters = list(lower)

df2 = pd.DataFrame()

for j in range(0, 5):
  rows=j
  for index, letter in enumerate(list_of_lowercase_letters):
      cols=index
      try:
         alpha_count = df1.iloc[:, j].value_counts()[{letter}]
         arr[rows,cols] = alpha_count
      except Exception as e:
         continue # continue to next iteration since this error is expected

df2 = pd.DataFrame(arr, 
                    index=['Letter1', 'Letter2', 'Letter3', 'Letter4', 'Letter5'],
                   columns=list(map(chr,range(ord('a'),ord('z')+1))) )

df2.head()
```
